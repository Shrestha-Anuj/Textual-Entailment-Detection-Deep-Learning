```python
import os
from datasets import load_dataset
import pandas as pd

os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"

ds = load_dataset("SetFit/mnli")
```

```python
# dictionary of dataset
ds
```

```python
# function for dataFrame to csv
def saveDF(data_frame,data_frame_name):
    
    data_frame.to_csv(data_frame_name + '.csv', index=False)
    return 0
```

```python
for enum in ds.keys():
    print(enum)
    dt_name = 'df_'+enum
    print(dt_name)
    dt = pd.DataFrame(ds[enum])
    saveDF(dt,dt_name)
    
```
