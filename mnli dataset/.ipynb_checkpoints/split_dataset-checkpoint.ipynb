# splitting the dataset

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# loading dataset for splitting
data_frame = pd.read_csv('df_train.csv')

data_frame
```

```python
# Rename multiple columns
def reName(data_frame):
    data_frame.rename(columns={
        "text1": "premise",
        "text2": "hypothesis"
    }, inplace=True)
    
    return 0
```

```python
reName(data_frame)
data_frame
```

```python
# Split into training and testing sets
train_dataframe, test_dataframe = train_test_split(data_frame, test_size=0.2, random_state=42)
```

```python
train_dataframe
```

```python
test_dataframe
```

```python
# rename the validation dataset
validation_data_frame = pd.read_csv('df_validation.csv')
reName(validation_data_frame)

validation_data_frame
```

```python
train_dataframe.to_csv("splitted_training_data.csv",index=False)
test_dataframe.to_csv("splitted_test_data.csv",index=False)
validation_data_frame.to_csv("validation_data.csv",index=False)
```
